#### Get data

```python
sample_data = [['HOLD_R', 3],
               ['HOLD_E', 5], ['HOLD_E_AFTER_R', 5], ['HOLD_R_E', 8], ['LATENCY_R_E', 3],
               ['HOLD_F', 4], ['HOLD_F_AFTER_E', 4], ['HOLD_E_F', 9], ['LATENCY_E_F', 2],
               ['HOLD_H_SHIFT', 9], ['HOLD_H_AFTER_F_SHIFT', 9], ['HOLD_F_H_SHIFT', 13], ['LATENCY_F_H_SHIFT', 6],
               ['HOLD_T', 7], ['HOLD_T_AFTER_H', 7], ['HOLD_H_T', 16], ['LATENCY_H_T', 3],
               ['HOLD_K', 3], ['HOLD_K_AFTER_T', 3], ['HOLD_T_K', 10], ['LATENCY_T_K', 4],
               ['HOLD_L', 4], ['HOLD_L_AFTER_K', 4], ['HOLD_K_L', 7], ['LATENCY_K_L', 3],
               ['HOLD_E', 4], ['HOLD_E_AFTER_L', 4], ['HOLD_L_E', 8], ['LATENCY_L_E', 4],
              ]
    
df = pd.DataFrame(sample_data, columns =['feature', 'value'])

# list with dataframe for every user
inputs = [df]
df
```

#### Calculate feature mean and standard deviation

```python
user_templates = []

# calculate for every user, put into single dataframe
for dataframe in inputs:
    template = pd.DataFrame()
    template['mean_'] = df.groupby('feature').mean()['value']
    template['std_'] = df.groupby('feature').std().fillna(0)['value']
    user_templates.append(template)
    
user_templates
```

#### Manhattan distance

```python
# Manhattan distance between test vector v and user template i
def manhattan(v, i):
    # need to adjust to something like dictionary so that manhattan compares between same features j and not index
    return sum(abs(feature-mean)/std for feature, mean, std in zip(
        v, user_templates[i]['mean'], user_templates[i]['std']))

manhattan ()
```
